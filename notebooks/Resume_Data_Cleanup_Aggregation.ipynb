### Mgr Profiles

```python
Mgr_Desc = df_Mgr[['description','user_id']].copy()
```

```python
Mgr_Desc.info()
```

```python
Mgr_Desc['user_id'] = Mgr_Desc['user_id'].astype(int)
Mgr_Desc['description'] = Mgr_Desc['description'].astype(str)
```

```python
Mgr_Desc_grouped = Mgr_Desc.groupby('user_id')
Mgr_Desc_Grp_Result = Mgr_Desc_grouped['description'].agg(lambda x: ','.join(str(v) for v in x))
```
